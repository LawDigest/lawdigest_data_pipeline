# DataFetcher 클래스 테스트

이 노트북은 `DataFetcher` 클래스의 모든 메서드가 정상적으로 동작하는지 확인하기 위한 테스트 코드를 포함합니다.
각 섹션은 특정 메서드에 대한 테스트를 수행합니다.

## 1. 환경 설정 및 초기화
`DataFetcher` 인스턴스를 생성합니다.

```python
import lawdigest_data_pipeline as data_pipeline
import pandas as pd
from datetime import datetime, timedelta

# Fetcher 인스턴스 생성 
fetcher = data_pipeline.DataFetcher()
print("DataFetcher initialized.")
```

## 2. 국회의원 데이터 수집 (`fetch_lawmakers_data`)
현직 국회의원 명단과 정보를 수집합니다.

```python
lawmakers_df = fetcher.fetch_lawmakers_data()

if lawmakers_df is not None and not lawmakers_df.empty:
    print(f"Successfully fetched {len(lawmakers_df)} lawmakers.")
    display(lawmakers_df.head(3))
else:
    print("Failed to fetch lawmakers data.")
```

## 3. 법안 데이터 수집 (`fetch_bills_data`)
특정 기간 동안 발의된 법안의 주요 내용을 수집합니다.

```python
# 최근 날짜로 테스트 
target_date = (datetime.now() - timedelta(days=7)).strftime('%Y-%m-%d')
today_date = datetime.now().strftime('%Y-%m-%d')

latest_bills_df = fetcher.fetch_bills_data(start_date=target_date, end_date=today_date, age="22")

if latest_bills_df is not None and not latest_bills_df.empty:
    print(f"Successfully fetched {len(latest_bills_df)} bills.")
    display(latest_bills_df.head(3))
else:
    print("No bills found for the specified period.")
```

```python
# API 동작 확인을 위해 정해진 날짜로 테스트
bills_df = fetcher.fetch_bills_data(start_date="2025-07-11", end_date="2025-07-11", age="22")

if bills_df is not None and not bills_df.empty:
    print(f"Successfully fetched {len(bills_df)} bills.")
    display(bills_df.head(3))
else:
    print("No bills found for the specified period.")
```

## 4. 공동발의자 정보 수집 (`fetch_bills_coactors`)
위에서 수집한 법안(`bills_df`)에 대한 공동발의자 정보를 수집합니다.

```python
# bills_df가 메모리에 있으면 자동으로 사용합니다.
coactors_df = fetcher.fetch_bills_coactors(df_bills=bills_df.head(5))

if coactors_df is not None and not coactors_df.empty:
    print(f"Successfully fetched coactors for {len(coactors_df)} bills.")
    display(coactors_df.head(3))
else:
    print("Failed to fetch coactors data.")
```

## 5. 의정활동 타임라인 수집 (`fetch_bills_timeline`)
법안 심사 진행 단계 등 의정활동 내역을 수집합니다.
- 해당 메서드는 데이터 수집에만 책임을 가짐. 컬럼 필터링은 WorkFlowManager 클래스의 update_bills_timeline 메서드에서 수행

```python
timeline_df = fetcher.fetch_bills_timeline(start_date='2025-07-11', end_date='2025-07-11', age='22')

if timeline_df is not None and not timeline_df.empty:
    print(f"Successfully fetched {len(timeline_df)} timeline events.")
    display(timeline_df.head(3))
else:
    print("No timeline data found.")
```

## 6. 법안 결과 데이터 수집 (`fetch_bills_result`)
법안의 최종 처리 결과(가결, 부결, 폐기 등)를 수집합니다.

```python
# 결과가 많이 나오는 본회의 날짜 주변으로 설정
result_df = fetcher.fetch_bills_result(start_date='2025-07-03', end_date='2025-07-03', age='22')

if result_df is not None and not result_df.empty:
    print(f"Successfully fetched {len(result_df)} bill results.")
    display(result_df.head(3))
else:
    print("No bill results found.")
```

## 7. 본회의 표결 데이터 수집 (`fetch_bills_vote`)
본회의에서 이루어진 표결 정보를 수집합니다.

```python
vote_df = fetcher.fetch_bills_vote(start_date='2025-07-03', end_date='2025-07-03', age='22')

if vote_df is not None and not vote_df.empty:
    print(f"Successfully fetched {len(vote_df)} votes.")
    display(vote_df.head(3))
else:
    print("No vote data found.")
```

## 8. 정당별 표결 결과 집계 (`fetch_vote_party`)
수집된 표결 데이터(`vote_df`)를 바탕으로 정당별 찬성표를 집계합니다.

```python
# vote_df가 메모리에 있으면 자동으로 사용합니다.
party_vote_df = fetcher.fetch_vote_party(df_vote=vote_df.head(5))

if party_vote_df is not None and not party_vote_df.empty:
    print(f"Successfully aggregated party votes for {len(party_vote_df)} entries.")
    display(party_vote_df)
else:
    print("Failed to aggregate party votes.")
```

## 9. 대안 법안 정보 수집 (`fetch_bills_alternatives`)
대안반영폐기된 법안들의 대안 법안 정보를 수집합니다.
테스트를 위해 대안이 포함될 가능성이 높은 법안 목록을 먼저 확보합니다.

```python
# 대안반영폐기된 법안이 포함된 데이터를 수집 (예: 2024년 2월)
alt_bills_df = fetcher.fetch_bills_data(start_date="2025-07-11", end_date="2025-07-11", age="22")

# 대안 수집 실행
alternatives_df = fetcher.fetch_bills_alternatives(df_bills=alt_bills_df)

if alternatives_df is not None and not alternatives_df.empty:
    print(f"Successfully fetched {len(alternatives_df)} alternative bill relations.")
    display(alternatives_df.head(3))
else:
    print("No alternative bills found.")
```
